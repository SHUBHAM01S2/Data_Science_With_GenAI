### 6. How do we handle categorical variables in Machine Learning? What are the common techniques?
    - Categorical variables are variables that can take on a limited, fixed number of values.
    The common techniques for handling categorical variables in machine learning include:
    1. Label Encoding: This technique assigns a unique integer to each category. For example, "red" = 0, "blue" = 1, "green" = 2.
    2. One-Hot Encoding: This technique creates a binary column for each category. For example, "red" = [1, 0, 0], "blue" = [0, 1, 0], "green" = [0, 0, 1].
    3. Binary Encoding: This technique converts categories into binary numbers and then splits the digits into separate columns.
    4. Target Encoding: This technique replaces each category with the mean of the target variable for that category.

### 25. Explain data encoding?
    - data encoding is the process of converting categorical data into a numerical format that can be used by machine learning algorithms. This is important because most algorithms work with numerical data and cannot handle categorical data directly. There are several methods for encoding categorical data, including:
    1. Label Encoding: This method assigns a unique integer to each category in the data. For example, if we have a column with categories 'red', 'green', and 'blue', label encoding would convert them to 0, 1, and 2 respectively. This method is simple but can introduce ordinal relationships that may not exist in the data.
    2. One-Hot Encoding: This method creates a new binary column for each category in the data. For example, if we have a column with categories 'red', 'green', and 'blue', one-hot encoding would create three new columns: 'is_red', 'is_green', and 'is_blue'. Each row would have a 1 in the column corresponding to its category and 0s in the others. This method is useful for nominal data where there is no ordinal relationship between categories.
